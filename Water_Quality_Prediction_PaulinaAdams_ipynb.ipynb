```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/water_potability.csv')
```

```python
#Hay valores nulos en este dataframe?
df.isnull().values.any()
```

```python
#Contar valores nulos por columna
df.isnull().sum()
print("\nConteo de valores nulos en todo el dataframe:", df.isnull().sum().sum())
print("\nPorcentaje de valores nulos por columna:")
print(df.isnull().sum() / len(df) * 100)
```

```python
#Eliminar datos nulos
df.dropna(inplace=True)
print("\nConteo de valores nulos en todo el dataframe:", df.isnull().sum().sum())
```

```python
df.info()
```

```python
#Analisis Exploratorio de Datos (EDA)
#Columnas numericas que representas variables quimicas
variables_quimicas = ['ph', 'Hardness', 'Solids', 'Chloramines', 'Sulfate', 'Conductivity', 'Organic_carbon', 'Trihalomethanes', 'Turbidity']

#Histogramas para cada variable
plt.figure(figsize=(12, 8))

for i, variable in enumerate(variables_quimicas, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[variable], kde=True)
    plt.title(f'Histograma de {variable}')
    plt.xlabel(variable)
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
